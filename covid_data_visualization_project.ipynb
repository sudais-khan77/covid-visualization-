```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=pd.read_csv("countries.csv")
y=x.sample(50)
sns.barplot(x="Country",y="Confirmed",data=y,hue="Country",ci=None,linewidth=1,edgecolor="white")
plt.tick_params(axis='x',labelrotation=90)
plt.title("(confirmed cases per country)",fontweight='bold',color='tomato',fontstyle='italic')
plt.ylabel("confirmed cases",fontweight='bold',color='tomato',fontstyle='italic')
plt.xticks(rotation='vertical')
plt.xlabel("countries",fontweight='bold',color='tomato',fontstyle='italic')
plt.style.use('dark_background')
plt.show()
```

```python
fig,axis=plt.subplots(1,3,figsize=(30,5))
axis[0].bar(y['Country'],y['Confirmed'])
axis[0].tick_params(axis='x',labelrotation=90)
axis[0].set_ylabel("confirmed cases")
axis[1].bar(y['Country'],y['Recovered'])
axis[1].set_ylabel("recovered cases")
axis[1].tick_params(axis='x',labelrotation=90)
axis[2].bar(y['Country'],y['Deaths'])
axis[2].tick_params(axis='x',labelrotation=90)
axis[2].set_ylabel("total cases")
plt.title("(confirmed cases, recovered cases and total cases per country)",fontweight='bold',fontstyle='italic',color='tomato')
plt.show()
```

```python
file=pd.read_csv("time.csv")
data=file.sample(40)
#print(data)
plt.plot(data['Date'],data['Deaths'],marker='d',markerfacecolor="orange",markersize=5)
plt.tick_params(axis='x',labelrotation=90)
plt.ylabel("total deaths")
plt.xlabel("date")
plt.figsize=(20,10)
plt.title("total deaths per date",fontweight='bold',color='tomato',fontstyle='italic')
plt.show()
```

```python
data=pd.read_csv("time.csv")
data=data.sample(50)
sns.pointplot(data=data,x='Country/Region',y='Deaths',ci=None,linestyle="dotted",marker='o',color='tomato',markerfacecolor='white',
              markersize=10,markeredgecolor="black")
plt.tick_params(axis='x',labelrotation=90)
plt.ylabel("total deaths")
plt.title("camparison of deaths between countries",fontweight='bold',color='tomato',fontstyle='italic')
plt.tick_params(axis='x',labelrotation=90)
plt.show()
```

```python
data=pd.read_csv("countries.csv")
data=data.sample(20)
plt.plot(data['Date'],data['Deaths'],marker='d',markerfacecolor="orange",markersize=5,
         color="tomato")
plt.ylabel("total deaths")
plt.xlabel("date")
plt.tick_params(labelrotation=90)
plt.figsize=(20,10)
plt.plot(data['Date'],data['Recovered'],color="yellow",marker='d',
         markerfacecolor="orange",markersize=5)
plt.fill_between(data['Date'],data['Deaths'],data['Recovered'],color='grey',alpha=0.3,
                 where=data['Deaths']>data['Recovered'],interpolate=True) #interpolate use kegi che kala
plt.fill_between(data['Date'],data['Deaths'],data['Recovered'],color='red',alpha=0.3, #condition na apply kegi
                 where=data['Deaths']<data['Recovered'],interpolate=True)
plt.legend(loc='best')
plt.title("total deaths and recovered cases camparison",fontweight='bold',color='tomato',fontstyle='italic')
plt.show()

```

```python
data=pd.read_csv("time.csv")
data=data.sample(15)
meow=plt.violinplot(data['Confirmed'],showmeans=True,showmedians=True)
plt.xlabel("confirmed cases",fontstyle='italic',color='tomato')
plt.ylabel("density",fontweight='bold',fontstyle='italic',color='tomato')
for body in meow['bodies']:
 body.set_facecolor="tomato"
 body.set_edgecolor="yellow"
 body.set_linewidth=5
 body.set_linestyle="dotted"
 meow["cmedians"].set_linestyle="dotted"
 meow["cmedians"].set_color="red"
 meow["cmeans"].set_linestyle="dotted"
 meow["cmeans"].set_color="red"
plt.show()
```

```python
data=pd.read_csv('us_deaths.csv')
data=data.sample(20)
plt.boxplot(data['Case'],patch_artist=True,boxprops={"facecolor":'tomato'}
,whiskerprops={"color":"yellow","linestyle":"dotted","linewidth":5},
            capprops={"color":"yellow","linestyle":"solid","linewidth":5})
plt.title("total cases in us ",fontstyle='italic',color='tomato',fontweight='bold')
plt.style.use("dark_background")
plt.show()
```

```python
data=pd.read_csv('countries.csv')
data=data.sample(50)
x=data.pivot(index='Date',columns='Country',values='Confirmed')
sns.heatmap(x,annot=True,cmap='summer')
plt.style.use('dark_background')
plt.xlabel("countries",fontweight='bold',color='tomato',fontstyle='italic')
plt.ylabel("date",fontweight='bold',color='tomato',fontstyle='italic')
plt.title("(confirmed cases per country and date)",fontweight='bold',color='tomato',fontstyle='italic')
plt.show()

```

```python
data=pd.read_csv('countries.csv')
data=data.sample(20)
sns.scatterplot(data=data,x='Deaths',y='Recovered',hue='Country',marker='d')
plt.xlabel("(total deaths)",fontweight='bold',color='tomato',fontstyle='italic')
plt.ylabel("(total recovered)",fontweight='bold',color='tomato',fontstyle='italic')
plt.legend(loc=(1,0))
plt.title("(number of deaths and recovered cases per country)",fontweight='bold',color='tomato',fontstyle='italic')
plt.show()
```

```python
data=pd.read_csv('us_deaths.csv')
data=data.sample(10)
plt.pie(data['Case'],labels=data['Province/State'],autopct='%.2f%%',shadow=True)
plt.title("total cases in us per state",fontweight='bold',color='tomato',fontstyle='italic')
plt.show()
```
